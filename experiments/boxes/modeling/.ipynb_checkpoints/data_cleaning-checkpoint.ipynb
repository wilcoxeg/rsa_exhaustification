```python
import json
import pandas as pd
import numpy as np
```

```python
with open("model_results.txt", "r") as inf:
    data = inf.read()
        
data = json.loads(data)
probs = []
for d1 in data:
    for d2 in d1:
        for d3 in d2:
            for d4 in d3:
                probs.append(d4)
print(len(probs))

# Both -- both, single, exhaust
# Not -- exhaust, single, both

print([x for x in probs[1]])

plist = []
for x in probs:
    params = x[0].split("_")
    alpha = float(params[1])
    c1 = float(params[2])
    c2 = float(params[3])
    listnr_low = x[1]['probs'][1] #Value listener assigns AB world in low prob condition
    listnr_high = x[2]['probs'][1] #Value listener assigns AB world in high prob condition
    # AB world
    ab_both = x[3]['probs'][0] #'ab' utterance
    ab_bare = x[3]['probs'][1] + x[3]['probs'][2] # sum of 'a' and 'b' utterance
    ab_exhaust = 0
    # A world (low prob)
    low_both = 0
    low_bare = x[4]['probs'][1]
    low_exhaust = x[4]['probs'][0]
    # B world (high prob)
    high_both = 0
    high_bare = x[5]['probs'][1]
    high_exhaust = x[5]['probs'][0]
    
    plist.append((alpha, c1, c2, listnr_low, listnr_high, ab_both, ab_bare, ab_exhaust,
                 low_both, low_bare, low_exhaust, high_both, high_bare, high_exhaust))


#df = pd.DataFrame(plist, columns=["alpha", "c1", "c2", "lstnr-low", "lstnr-high"])

df = pd.DataFrame(plist, columns=["alpha", "c1", "c2", "lstnr-low", "lstnr-high",
                                    "AB-both", "AB-bare", "AB-exhaust",
                                    "A-both", "A-bare", "A-exhaust",
                                    "B-both", "B-bare", "B-exhaust"])


#df.to_csv('model_spkr-lstnr-results.csv', index=False)
df.head()
```

```python
# Calculate the MSE

num_samples = 8820
def mse(array1, array2):
    return np.square(np.array(array1) - np.array(array2)).mean(axis=0)

### The numbers here are the human results taken from the boxes experiment!

#Low prob, High Prob
human_listener = np.array([[0.045, 0.22],]*num_samples).transpose()

# A = low, B = high
#AB-bare, AB-both, AB-exhaust, A-bare, A-both, A-exhaust, B-both, B-bare, B-exhaust
human_speaker = np.array([[0.15, 0.79, 0.06, 0.5, 0.01, 0.49, 0.33, 0.05, 0.61],]*num_samples).transpose()
human_spkr_listnr = np.array([[0.045, 0.22, 0.15, 0.79, 0.06, 0.5, 0.01, 0.49, 0.33, 0.05, 0.61],]*num_samples).transpose()


df["lstnr_mse"] = mse(np.array([df["lstnr-low"],df["lstnr-high"]]), human_listener)
df["spkr_mse"] = mse(np.array([df["AB-both"],df["AB-bare"],df["AB-exhaust"],df["A-both"],df["A-bare"],df["A-exhaust"],df["B-both"],df["B-bare"],df["B-exhaust"]]), human_speaker)
df["lstnr_spkr_mse"] = mse(np.array([df["lstnr-low"],df["lstnr-high"],df["AB-both"],df["AB-bare"],df["AB-exhaust"],df["A-both"],df["A-bare"],df["A-exhaust"],df["B-both"],df["B-bare"],df["B-exhaust"]]), human_spkr_listnr)

df = df.sort_values(by='lstnr_mse', ascending=True)
df.head()

df.to_csv('fit_results_nonmodal.csv', index=False)
```
